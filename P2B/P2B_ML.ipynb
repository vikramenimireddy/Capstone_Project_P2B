# Linear regression model

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import seaborn as sns
# function of linear regression
def StudentReg(ages_train, net_worth_train):
  from sklearn.linear_model import LinearRegression
  reg = LinearRegression()
  reg.fit(ages_train, net_worth_train)
  return reg
#random() function is used to generate random numbers in Python
#Seed is used in the generation of a pseudo-random encryption key.
#Also seed function is used to generate same random numbers again and again and 
#simplifies algorithm testing process.
random.seed(42) 
np.random.seed(42)
ages = []
for ii in range(100):
  ages.append(np.random.randint(20,65))

net_worth = [ii * 6.25 + np.random.normal(scale=40) for ii in ages]
# reshaping data to fit into model
ages = np.reshape(np.array(ages),(len(ages),1))
net_worth = np.reshape(np.array(net_worth),(len(net_worth),1))
# dividing data set into train & test data
from sklearn.model_selection import train_test_split
ages_train, ages_test , net_worth_train, net_worth_test = train_test_split(ages,net_worth)
# Linear regression model(function calling) calling and passing tarining data to model
reg1 = StudentReg(ages_train, net_worth_train)
#Checking slope and intercept of the trained model
print("Coefficient",reg1.coef_)
print("Slope",reg1.intercept_)
#Calculating efficiency. It internally calculates y_pred again and gives the 
#efficiency
print('Training data efficiency: ',reg1.score(ages_train, net_worth_train)*100)
print('Testing data efficiency: ',reg1.score(ages_test, net_worth_test)*100)
nw_pred = reg1.predict(ages_test)
# plotting 
  plt.figure(figsize = (10,10))
  plt.scatter(ages,net_worth,color = 'Green')
  plt.scatter(ages_train,net_worth_train,color='b',label = 'training data')
  plt.scatter(ages_test,net_worth_test,color='r',label = 'testing dta')
  plt.plot(ages_test,nw_pred,color='r',label= 'Linear Regression Straight line of Test data')
  plt.legend(loc ="upper left");
  plt.xlabel('age')
  plt.ylabel('Net Worth')
  plt.title('Age - Networth Linear Regression Model')
  plt.savefig('AgeNetworthLinearRegressionModel.png',transparent = True)
```

# Random Forest Regressor and Classifier

```python
# importing rnadom forest classifier or regressor with train_test_split and make classification and make regreession
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
```

Random Forest Regressor

```python
# loading random data set with 4 feature among which 2 are informative
X,y = make_regression(n_samples =100, n_features=4,n_informative=2,random_state=0,shuffle=False)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state =0 ,test_size = 0.15)
```

```python
reg = RandomForestRegressor(max_depth=2,random_state=0)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
# Mall customers clustering
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="notebook", palette="Spectral", style = 'darkgrid' ,font_scale = 1.5, color_codes=True)
customers = pd.read_csv('Mall_Customers.csv',index_col ='CustomerID' )
customers.info()
```

```python
customers.drop_duplicates(inplace=True)
```

```python
X = customers.iloc[:,[2,3]].values
```

```python
X
```

```python
# elbow method
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++',random_state=42)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
sns.lineplot(range(1,11),wcss,marker='o',color='red');
plt.title('Clusters')
plt.xlabel('number of clusters')
plt.ylabel('wcss')
```

```python
kmeans  =KMeans(n_clusters=5,init = 'k-means++',random_state=42)
y_kmeans = kmeans.fit_predict(X)
```
